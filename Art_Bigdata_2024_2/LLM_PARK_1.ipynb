```python
from transformers import pipeline

#텍스트를 생성, 파이프라인 설정
generator = pipeline('text-generation', model = 'gpt2')

#텍스트 입력 및 결과 생성
result = generator("Once upon a time")
print(result)
```

```python
from transformers import pipeline

#텍스트를 생성, 파이프라인 설정
qa_pipeline = pipeline('question-answering')

#텍스트 입력 및 결과 생성
result = qa_pipeline({"question" : "What is the capital of France?",
                      'context' : 'France is a country in Europe. Its capital is France'})

print(result)
```

```python
from transformers import pipeline

#텍스트를 생성, 파이프라인 설정
translator = pipeline('translation_en_to_fr')

#텍스트 입력 및 결과 생성
text = 'Hugginf face is creating amazing tools for developers and researchers.'
result = translator(text)

print(result)

#No model was supplied, defaulted to google-t5/t5-base and revision 686f1db (https://huggingface.co/google-t5/t5-base).
#모델 지정 X, 디폴트값으로 설정이 됨.
```

```python
from transformers import pipeline

#텍스트를 생성, 파이프라인 설정
summarizer = pipeline('summarization')

#텍스트 입력 및 결과 생성
result = summarizer("""Transformer research has led to groundbreaking achievements in artificial intelligence (AI) and natural language processing (NLP). Here are some key accomplishments of Transformer research:

Revolution in NLP: Unlike previous models like RNNs (LSTM, GRU), Transformers can handle long-range dependencies efficiently and process tasks in parallel. This led to significant improvements in tasks such as translation, summarization, and question answering. Notable models like BERT, GPT, and T5 are based on the Transformer architecture.

Self-Attention Mechanism: Transformers introduced the self-attention mechanism, allowing the model to learn how each word in a sentence relates to others. This greatly improved the model’s ability to understand context and boosted performance in NLP tasks.

Pretraining and Transfer Learning: Transformer-based models are pretrained on large-scale text data and can be fine-tuned for specific downstream tasks. This has become a standard approach in NLP, with models like BERT and GPT excelling at a wide range of applications, often requiring less task-specific data.

Expansion to Various Fields: Beyond NLP, Transformers have been applied in image processing (Vision Transformers, ViT), speech recognition, time series analysis, and even game-playing (AlphaStar). This demonstrates the flexibility of the Transformer architecture.

Large-Scale Models: Models like GPT-3, with billions of parameters, have shown remarkable capabilities in generating human-like text, translating languages, and performing other NLP tasks. This proves that the Transformer structure scales well and maintains high performance.

"Attention Is All You Need": The 2017 paper introduced the idea of replacing sequential structures with parallel processing, leading to improved efficiency and performance. This paper marked the foundation of Transformer research and has become one of the most cited papers in AI.

Transformer research represents a major turning point in AI, driving advancements across multiple domains.""")

print(result)
```
