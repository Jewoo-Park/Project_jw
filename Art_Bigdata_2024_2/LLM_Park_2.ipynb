```python
#step2 Hugging Face에서 GPT모델 선택 및 다운로드
from transformers import AutoModelForCausalLM, AutoTokenizer

#모델과 토크나이저 생성
model_name = 'TurkuNLP/gpt3-finnish-large' #Hugging Face에서 사용할 모델 이름 meta-llama/Llama-3.2-1B 등
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
#Step3 챗봇 함수 작성
import torch
def generate_response(prompt):
    # 사용자 입력을 토큰화
    inputs = tokenizer(prompt, return_tensors="pt")
    # 모델로 텍스트 생성
    outputs = model.generate(
        inputs["input_ids"],
        max_length=100,      # 최대 생성 길이 설정
        num_return_sequences=1,  # 한 개의 응답만 생성
        no_repeat_ngram_size=2,  # 반복 방지
        top_p=0.9,             # 상위 확률 샘플링
        do_sample=True #샘플링 활성화
    )
#토큰을 텍스트로 디코딩
    response = tokenizer.decode(outputs[0], skip_special_tokens = True)
    return response
```

```python
#Step4 대화 텍스트
print("챗봇과 대화를 시작해 보세요! '종료'라고 입력하면 종료됩니다.")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["종료", "exit"]:
        print("대화를 종료합니다. 감사합니다.")
        break

    response = generate_response(user_input)  # 들여쓰기 수정
    print("Chatbot:", response)
```
